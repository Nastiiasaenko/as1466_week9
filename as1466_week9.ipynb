#### Step 1. Imports and dataset download

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import shutil
```

```python
# Replace "github.com" with "raw.githubusercontent.com" and remove "/blob/"
path = r"https://raw.githubusercontent.com/nogibjj/as1466_week9/main/life%20expectancy%202.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df = df.dropna(axis=1, how='all')

# Display the correlation matrix
corr_matrix = df.corr()

# Plot the correlation matrix heatmap using seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='CO2', y='Life Expectancy World Bank', hue='Region', data=df, palette='viridis')

# Set plot title and labels
plt.title('Scatter Plot: CO2 vs Life Expectancy')
plt.xlabel('CO2 Emissions')
plt.ylabel('Life Expectancy (World Bank)')

# Add legend
scatter_plot.legend(title='Country Name', bbox_to_anchor=(1, 1), loc='upper left')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='Region', y='Life Expectancy World Bank', data=df, palette='Set3')

# Set plot title and labels
plt.title('Boxplot: CO2 Emissions by Region')
plt.xlabel('Region')
plt.ylabel('CO2 Emissions')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 8))
box_plot = sns.boxplot(x='Year', y='Life Expectancy World Bank', data=df, palette='Set3')

# Set plot title and labels
plt.title('Boxplot: CO2 Emissions by Region')
plt.xlabel('Region')
plt.ylabel('CO2 Emissions')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
